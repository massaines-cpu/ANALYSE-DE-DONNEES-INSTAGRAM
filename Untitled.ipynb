<h2 style="color:pink; text-align:center; font-size:30px;">
QUESTION 1 : PRÉDIRE LE NIVEAU DE STRESS EN FONCTION DU TEMPS PASSÉ SUR INSTAGRAM</h2>

<h3 style="color:orange;">
On cherche un nombre donc on fait une régression linéaire</h3>

```python
import matplotlib.pyplot as plt
X = insta_data[['daily_active_minutes_instagram']]
y = insta_data['perceived_stress_score']
plt.scatter(X, y)
```

<h3 style="color:orange;">
Il faut prédire, donc on doit entraîner un modèle</h3>

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=1/3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
print(regressor.coef_)
print(regressor.intercept_)
```

```python
y_predict = regressor.predict(X_test)
```

<h3 style="color:orange;">
On calcule l'erreur moyenne (valeur absolue), on la met au carré, puis on fait sa racine carré, le r2_score c'est le pourcentage de la variance du modèle</h3>

```python
from sklearn import metrics
print('mean absolute error:', metrics.mean_absolute_error(y_test, y_predict))
print('mean squared error:', metrics.mean_squared_error(y_test, y_predict))
print('racine de mean squared error:', np.sqrt(metrics.mean_squared_error(y_test, y_predict)))
print('racine au carré:', metrics.r2_score(y_test, y_predict))
```

<h3 style="color:orange;">
Graphique des données prédites</h3>

```python
plt.scatter(X_test, y_test, color='blue', label='réel')
plt.scatter(X_test, y_predict, color='red', label='prédit')
plt.xlabel('minutes insta')
plt.ylabel('stress')
plt.legend()
plt.show()
```

On crée des tranches d'ages pour que ce soit plus simple</h3>

```python
bins = [0, 25, 35, 50, 100]
labels = ['18-25', '26-35', '36-50', '50+']
insta_data['tranche_age'] = pd.cut(insta_data['age'], bins=bins, labels=labels, include_lowest=True)
```

<h3 style="color:orange;">
Calcul de la moyenne des minutes passées sur instagram par tranche âge</h3>

```python

moyenne_minutes = insta_data.groupby('tranche_age', observed=True)['daily_active_minutes_instagram'].mean()

#graphique barres
#index= tranche dage
#values = la moyenne en fonction des tranches d'age
plt.bar(moyenne_minutes.index, moyenne_minutes.values, color='green')
plt.xlabel('tranche âge')
plt.ylabel('moyenne minutes instagram')
plt.title('temps moyen sur instagram par tranche âge')
plt.show()
```

<h3 style="color:orange;">
Définir X et y pour entraîner le modèle </h3>

```python
X = insta_data[['daily_active_minutes_instagram',
                'reels_watched_per_day',
                'stories_viewed_per_day',
                'sessions_per_day',
                'posts_created_per_week',
                'likes_given_per_day',
                'followers_count',
                'comments_written_per_day',
                'dms_sent_per_week',
                'time_on_reels_per_day',
                'time_on_feed_per_day',
                'average_session_length_minutes']]
y = insta_data['tranche_age']
```

<h3 style="color:orange;">
On entraîne le modèle avec un random forest de classification </h3>

```python

from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)

entrainement = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=0)
entrainement.fit(X_train, y_train)
```

<h3 style="color:orange;">
Mes pourcentages de prédictions </h3>

```python
from sklearn.metrics import classification_report

y_predict = entrainement.predict(X_test)
print(classification_report(y_test, y_predict))
```

<h3 style="color:orange;">
On fait une matrice de confusion pour mieux voir le niveau du modèle </h3>

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix

confusion = confusion_matrix(y_test, y_predict)
disp = ConfusionMatrixDisplay(confusion_matrix=confusion,
                            display_labels=['18-25', '26-35', '36-50', '50+'])
disp.plot(cmap='Blues')
plt.title('matrice de confusion')
plt.show()
```

<h3 style="color:orange;">Moyenne de création de post en fonction du type de créateur</h3>


```python
moyenne_creation_post = insta_data.groupby('type_createur', observed=True)['posts_created_per_week'].mean()

#graphique barres
#index= type de créateurs 'connus'
#values = la moyenne en fonction des contenus postés
plt.bar(moyenne_creation_post.index, moyenne_creation_post.values, color='green')
plt.xlabel('type de créateurs')
plt.ylabel('moyenne post crée')
plt.title('moyenne post crée en fonction de type createurs')
plt.show()
```

<h3 style="color:orange;">On définit les types de comportements pouvant potentiellement être influencés par le fait de beaucoup poster</h3>




<h4 style="color:lightblue;">On fait un boxplot car c'est des quartiles

```python
variables = ['daily_active_minutes_instagram',
             'sessions_per_day',
             'dms_received_per_week',
             'comments_written_per_day',
             'books_read_per_year',
             'self_reported_happiness',
             'perceived_stress_score',
             'sleep_hours_per_night']

fig, axes = plt.subplots(2, 4, figsize=(18, 10))
axes = axes.flatten()

for i, var in enumerate(variables):
    insta_data.boxplot(column=var, by='type_createur', ax=axes[i])
    axes[i].set_title(var)
    axes[i].set_xlabel('type créateur')

plt.suptitle('comportement selon le type de créateur')
plt.tight_layout()
plt.show()
```
