```python
import pandas as pd

insta_filepath = "instagram_users_lifestyle.csv"
insta_data = pd.read_csv(insta_filepath)
insta_data.head()
```

```python
## QUESTION 1 : PRÉDIRE LE NIVEAU DE STRESS  FONCTION DU TEMPS PASSÉ SUR INSTAGRAM
```

```python
import matplotlib.pyplot as plt
X = insta_data[['daily_active_minutes_instagram']]
y = insta_data['perceived_stress_score']
plt.scatter(X, y)
```

```python
#regression lineaire
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=1/3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
print(regressor.coef_)
print(regressor.intercept_)
```

```python
ordonnee = np.linspace(0, 600, 1000)
plt.scatter(X, y)
plt.plot(ordonnee, regressor.coef_[0]*ordonnee+regressor.intercept_, color='red')
plt.xlabel('Minutes instagram')
plt.ylabel('Stress')
```

```python
#entraitement
y_predict = regressor.predict(X_test)
```

```python
from sklearn import metrics
print('mean absolute error:', metrics.mean_absolute_error(y_test, y_predict))
print('mean squared error:', metrics.mean_squared_error(y_test, y_predict))
print('racine de mean squared error:', np.sqrt(metrics.mean_squared_error(y_test, y_predict)))
print('racine au carré:', metrics.r2_score(y_test, y_predict))
```

```python
#test avec des données temporelles pour prédire le stress
nouvelles_infos = [30, 250, 700]
nouveau_test = pd.DataFrame(nouvelles_infos, columns=['daily_active_minutes_instagram'])
y_nouveau = regressor.predict(nouveau_test)
print(y_nouveau)
```

```python
#graphique qui predit
plt.scatter(X_test, y_test, color='blue', label='réel')
plt.scatter(X_test, y_predict, color='red', label='prédit')
plt.xlabel('minutes insta')
plt.ylabel('stress')
plt.legend()
plt.show()
```

```python
## QUESTION 4 : QUELLE INFLUENCE A LE FAIT DE CREER BEAUCOUP DE POST SUR LE COMPORTEMENT DE L'UTILISATEUR (COMPARER LES GROS CREATEURS AVEC LES PETITS, PAR QUARTILE?)
```

```python
##comment reconnaitre les gros créateurs et les différencier des petits???
```

```python
#infos dont je pense avoir besoin pour les gros créateurs
# social_events_per_month
# travel_frequency_per_year
# dms_received_per_week
#followers_count
#following_count
# subscription_status
# uses_premium_features
# posts_created_per_week
```

```python
print(insta_data['posts_created_per_week'].describe())
```

```python
# 25% = 3  25% des utilisateurs créent 3 posts voir moins par semaine = petits créateurs
# 50% = 5  la moitié des utilisateurs créent 5 posts ou moins
# 75% = 8  75% des utilisateurs créent 8 posts ou moins
# donc les 25% restants créent plus de 8 = les gros créateurs
```

```python
#on cree les tranches de créateurs
```

```python
bins = [0, 3, 8, 30]
#0 cest le min, 30 c'est le max
labels = ['petits créateurs', 'regular créateurs', 'influenceurs de dingue']
insta_data['type_createur'] = pd.cut(insta_data['posts_created_per_week'], bins=bins, labels=labels, include_lowest=True)
#include_lowest=True jai du rajouter ca car ca me bloquait tout, ca veut dire que meme si une personne poste pas on la compte
```

```python
moyenne_creation_post = insta_data.groupby('type_createur', observed=True)['posts_created_per_week'].mean()

#graphique barres
#index= tranche dage
#values = la moyenne en fonction des tranche d'age
plt.bar(moyenne_creation_post.index, moyenne_creation_post.values, color='green')
plt.xlabel('type de créateurs')
plt.ylabel('moyenne post crée')
plt.title('moyenne post crée en fonction de type createurs')
plt.show()
```

```python
#variables comportement utiles :
X = insta_data[[
    'daily_active_minutes_instagram',
    'sessions_per_day',
    'likes_given_per_day',
    'comments_written_per_day',
    'dms_received_per_week',
    'followers_count',
    'perceived_stress_score',
    'sleep_hours_per_night'
]]
y = insta_data['type_createur']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)

entrainement = DecisionTreeClassifier()
entrainement.fit(X_train, y_train)
```

```python
y_predict = entrainement.predict(X_test)
print(classification_report(y_test, y_predict))
```

```python
#il a pratiquement une chance sur 2 de se tromper mais bon....ça marche
```

```python
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrix

confusion = confusion_matrix(y_test, y_predict)
disp = ConfusionMatrixDisplay(confusion_matrix=confusion,
                            display_labels=['petits créateurs', 'regular créateurs', 'influenceurs de dingue'])
disp.plot(cmap='Blues')
plt.title('matrice de confusion')
plt.show()
```

```python
# diagonale gauche haut a bas droit = bonne prediction
# 58 650 vrais petits créateurs classés comme "influenceurs de dingue"
# 17 188 vrais petits créateurs classés comme "regular"
# 60 583 vrais influenceurs classés comme "petits créateurs"
# du coup les petits créateurs sont confondus avec les gros et vice versa, ca veut dire qu'ils ont les memes comportements...
```

```python
#graphique donnees predites
#on dirait les memes mais c'est qu'en fait il confond les gros et les petits, donc y a peut etre des petits qui ont ete consideres comme gros et des gros comme des petits. en gros comme y a des erreurs dans les deux sens ça s'annule
plt.figure(figsize=(6, 5))
sns.countplot(x=y_predict, color='red', order=labels)
plt.title('Prédit')
plt.show()
```
