# QUESTION 1 : PRÉDIRE LE NIVEAU DE STRESS  FONCTION DU TEMPS PASSÉ SUR INSTAGRAM


```python
import matplotlib.pyplot as plt
X = insta_data[['daily_active_minutes_instagram']]
y = insta_data['perceived_stress_score']
plt.scatter(X, y)
```

regression lineaire

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=1/3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
print(regressor.coef_)
print(regressor.intercept_)
```

```python
ordonnee = np.linspace(0, 600, 1000)
plt.scatter(X, y)
plt.plot(ordonnee, regressor.coef_[0]*ordonnee+regressor.intercept_, color='red')
plt.xlabel('Minutes instagram')
plt.ylabel('Stress')
```

entraitement

```python
y_predict = regressor.predict(X_test)
```

pas très utile de les print mais je voulais voir

```python
from sklearn import metrics
print('mean absolute error:', metrics.mean_absolute_error(y_test, y_predict))
print('mean squared error:', metrics.mean_squared_error(y_test, y_predict))
print('racine de mean squared error:', np.sqrt(metrics.mean_squared_error(y_test, y_predict)))
print('racine au carré:', metrics.r2_score(y_test, y_predict))
```

graphique qui predit

```python
plt.scatter(X_test, y_test, color='blue', label='réel')
plt.scatter(X_test, y_predict, color='red', label='prédit')
plt.xlabel('minutes insta')
plt.ylabel('stress')
plt.legend()
plt.show()
```

## Créer les catégories de créateurs : petits,moyen,grand

### On regarde ou se situent les personnes en fonction de leurs post créés

```python
print(insta_data['posts_created_per_week'].describe())
```

25% = 3 --> 25% des utilisateurs créent 3 posts voir moins par semaine = petits créateurs

50% = 5 --> la moitié des utilisateurs créent 5 posts ou moins

75% = 8 --> 75% des utilisateurs créent 8 posts ou moins

donc les 25% restants créent plus de 8 = les gros créateurs

```python
bins = [0, 3, 8, 30]
#0 cest le min, 30 c'est le max
labels = ['petits créateurs', 'regular créateurs', 'influenceurs de dingue']
insta_data['type_createur'] = pd.cut(insta_data['posts_created_per_week'], bins=bins, labels=labels, include_lowest=True)
#include_lowest=True jai du rajouter ca car ca me bloquait tout, ca veut dire que meme si une personne poste pas on la compte
```

### On a donc les petits créateurs, les regular et les influenceurs de dingue

## Moyenne de création de post en fonction du type de créateur

```python
moyenne_creation_post = insta_data.groupby('type_createur', observed=True)['posts_created_per_week'].mean()

#graphique barres
#index= type de créateurs 'connus'
#values = la moyenne en fonction des contenus postés
plt.bar(moyenne_creation_post.index, moyenne_creation_post.values, color='green')
plt.xlabel('type de créateurs')
plt.ylabel('moyenne post crée')
plt.title('moyenne post crée en fonction de type createurs')
plt.show()
```

### On définit les types de comportements pouvant potentiellement être influencés par le fait de beaucoup poster

### On fait un boxplot car c'est des quartiles
J'ai choisi ces variables par logique et bon sens OK



```python
variables = ['daily_active_minutes_instagram',
             'sessions_per_day',
             'hobbies_count',
             'books_read_per_year',
             'exercise_hours_per_week',
             'self_reported_happiness',
             'perceived_stress_score',
             'sleep_hours_per_night']

fig, axes = plt.subplots(2, 4, figsize=(18, 10))
axes = axes.flatten()

for i, var in enumerate(variables):
    insta_data.boxplot(column=var, by='type_createur', ax=axes[i])
    axes[i].set_title(var)
    axes[i].set_xlabel('type créateur')

plt.suptitle('comportement selon le type de créateur')
plt.tight_layout()
plt.show()
```

# Résultat : Les influenceurs passent + de temps sur 1sta, mettent + de like, écrivent + de commentaires, reçoivent + de dm, sont + stressés, ils dorment bien la nuit par contre, ils sont moins heureux
