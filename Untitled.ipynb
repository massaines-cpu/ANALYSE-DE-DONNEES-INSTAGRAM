```python
import pandas as pd

insta_filepath = "instagram_users_lifestyle.csv"
insta_data = pd.read_csv(insta_filepath)
insta_data.head()
```

```python
list(insta_data.columns)
```

```python
#QUESTION 1 : PRÉDIRE LE NIVEAU DE STRESS EN FONCTION DU TEMPS PASSÉ SUR INSTAGRAM
import matplotlib.pyplot as plt
X = insta_data[['daily_active_minutes_instagram']]
y = insta_data['perceived_stress_score']
plt.scatter(X, y)
```

```python
#regression lineaire
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=1/3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
```

```python
regressor.fit(X_train, y_train)
```

```python
print(regressor.coef_)
print(regressor.intercept_)
```

```python
ordonnee = np.linspace(0, 600, 1000)
plt.scatter(X, y)
plt.plot(ordonnee, regressor.coef_[0]*ordonnee+regressor.intercept_, color='red')
plt.xlabel('Minutes instagram')
plt.ylabel('Stress')
```

```python
#entraitement
y_predict = regressor.predict(X_test)
```

```python
from sklearn import metrics
print('mean absolute error:', metrics.mean_absolute_error(y_test, y_predict))
print('mean squared error:', metrics.mean_squared_error(y_test, y_predict))
print('racine de mean squared error:', np.sqrt(metrics.mean_squared_error(y_test, y_predict)))
print('racine au carré:', metrics.r2_score(y_test, y_predict))
```

```python
#test avec des données temporelles pour prédire le stress
nouvelles_infos = [30, 250, 700]
nouveau_test = pd.DataFrame(nouvelles_infos, columns=['daily_active_minutes_instagram'])
y_nouveau = regressor.predict(nouveau_test)
y_nouveau
```

```python
#graphique qui predit
plt.scatter(X_test, y_test, color='blue', label='réel')
plt.scatter(X_test, y_predict, color='red', label='prédit')
plt.xlabel('Stress')
plt.ylabel('Minutes Instagram')
plt.legend()
plt.show()
```
